```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# single feature data
data = pd.read_csv("../Data/train_sample_rand_1000.csv")
```

```python
data_arr = data.to_numpy()
X = data_arr[:, 1].reshape(-1, 1)
y = data_arr[:, 2].reshape(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
k = 50
knn = KNeighborsRegressor(n_neighbors=k)
knn.fit(X_train, y_train)
```

```python


x_line = np.arange(min(X_train), max(X_train) + 1).reshape(-1, 1)

y_pred = knn.predict(x_line)
```

```python
plt.title("Scatter plot of sample training data with overlayed KNN regression")
plt.scatter(X, y, s=1)
plt.xlabel('acoustic signal')
plt.ylabel('time to failure')
plt.plot(x_line, y_pred, c="red")
plt.show()
```

```python
y_pred_test = knn.predict(X_test)
mean_squared_error(y_pred_test, y_test)
```
